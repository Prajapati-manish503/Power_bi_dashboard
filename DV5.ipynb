```python
import pandas as pd

# Data create karna
data = {
    "ID": [101, 102, 103, 104],
    "Name": ["Manish", "Rahul", "Sneha", "Priya"],
    "City": ["Mumbai", "Delhi", "Pune", "Nagpur"],
    "Salary": [50000, 60000, 55000, 65000],
    "Email": [
        "manish@gmail.com",
        "rahul@gmail.com",
        "sneha@gmail.com",
        "priya@gmail.com"
    ]
}

# DataFrame create karna
df = pd.DataFrame(data)

# Output display
print(df)
```

```python
#Convert Name to Uppercase
df["Name_Upper"] = df["Name"].str.upper()
print(df)
```

```python
#Convert City to Lowercase
df["City_Lower"] = df["City"].str.lower()
```

```python
df["Name_Length"] = df["Name"].str.len()
print(df)
```

```python
#Extract Email Username
#manish@gmail.com â†’ manish

df["Username"] = df["Email"].str.split("@").str[0]
print(df)
```
